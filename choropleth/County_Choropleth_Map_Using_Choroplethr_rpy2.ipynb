<center><h2>Creating Choropleth Map using Choroplethr via rpy2</h2></center>

This is a jupyter notebook that explains how to make a choropleth map using R's choroplethr package from within Python via rpy2 interface.  rpy2 is a Python package that allows you to run R code from within Python.  Why am I doing this?  Currently, there isn't a library or package in Python world that makes choropleth maps as easily as R's choroplethr package.  A choropleth map is a geographical visualization map in which the intensity of the colored shaded regions corresponds to a metric or measurement.

In this example, I will be generating a choropleth map representing unemployment rates by US counties.  The data can be obtained [here](http://www.bls.gov/lau/#tables) in the "COUNTY DATA" section.  The raw data that I am working with is [here](http://www.bls.gov/lau/laucntycur14.txt).

If you look at the raw data, you can see that I need to ignore or skip the first 6 rows of data which means I will have to make my own set of column names.  The column data is separated or delimited by a "|" character.  Reading the code below, you should be able to guess what it does:

```python
import pandas as pd

df = pd.read_csv('http://www.bls.gov/lau/laucntycur14.txt', header=None, skiprows=6, sep='|',
                names=['LAUS','FIPS_STATE','FIPS_COUNTY','AREA_TITLE','PERIOD','CIV_LABOR_FORCE','EMPLOYED',
                       'UNEMPL_LEVEL','UNEMPL_RATE'],
                dtype={'FIPS_STATE':str, 'FIPS_COUNTY':str})
```

```python
df.head()
```

Per choroplethr [documentation](http://cran.r-project.org/web/packages/choroplethr/vignettes/c-county-choropleth.html) under "Data Requirements", we need to create a "region" column which is the FIPS state code and FIPS county code meshed together.  But first, we need to remove leading and trailing whitespaces and leading zeros.

```python
df['FIPS_STATE'] = df['FIPS_STATE'].str.strip()         # remove/strip leading or trailing whitespaces
df['FIPS_STATE'] = df['FIPS_STATE'].str.lstrip('0')     # left strip / remove leading zeros
df['FIPS_COUNTY'] = df['FIPS_COUNTY'].str.strip()
```

```python
df['region'] = df.FIPS_STATE + df.FIPS_COUNTY           # concatenate or mesh together the state and county FIPS code
df['value'] = df['UNEMPL_RATE']                         # create a column called "value" containing unemployment rates
```

```python
df.head()                                               # Let's see what the first 5 rows of our data looks like
```

### The description of the data set from the web site says it has data from Mar'13 through Apr'15:

```python
df.PERIOD.value_counts()
```

### We can see from above that for each time period, there are 3,219 rows of data, with the exception of "Apr-15" data.

### Let's just obtain the most recent data and we also need to exclude Puerto Rico which has a FIPS code of 72 

```python
criteria1 = df['PERIOD'].str.contains('Mar-15', na=False)
criteria2 = df['FIPS_STATE'] != '72'
recent = df[criteria1 & criteria2]
```

### Let's double-check if our data only has Mar'15 data

```python
recent.head()
```

```python
recent.PERIOD.value_counts()
```

Alright, so our data set consists of just Mar'15 data and contains 3,141 rows

### Our choropleth data doesn't need all those columns, so we'll just grab the ones that matter

```python
choropleth_data = recent[['region','value','AREA_TITLE']]
```

```python
choropleth_data.head()
```

### Load the rpy2 extension so that we can execute R code from within Python

```python
# load the rpy2 extension
%load_ext rpy2.ipython
```

### Send the choropleth_data to the R environment

```python
%R -i choropleth_data
```

### Now load the choropleth package, convert the Python data frame to R data frame and make the county choropleth map

```python
%%R -w 800 -h 500 -u px
library(choroplethr)
choropleth_data = as.data.frame(choropleth_data)
county_choropleth(choropleth_data)
```

### Error message above is telling us that it doesn't like that one of our columns is a "factor" data type.  Not sure why R wasn't smart enough to import the columns to the proper data type.  But anyways, I'll convert the region column as numeric type:

```python
%%R -w 800 -h 400 -u px
library(choroplethr)
choropleth_data = as.data.frame(choropleth_data)
choropleth_data$region = as.numeric(as.character(choropleth_data$region))
county_choropleth(choropleth_data, title = "Unemployment Rate by US Counties (March 2015)",
                 legend = "Unemployment Rate")
```

### We're done!  As you can see from this example, the more tedious part was not the actual creation of the choropleth map, but the data cleaning and transformation.
