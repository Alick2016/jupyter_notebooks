```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
import pandas as pd
from bokeh.charts import Histogram, output_notebook, show, hplot
from bokeh.sampledata.autompg import autompg as df
from bokeh.models import HoverTool
output_notebook()
pd.set_option("display.max_rows",1000)
```

```python
df.head()
```

```python
# Create custom hover tool output
hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("data (x,y)", "($x, $y)")  # for histogram it is "data (x,y)"
        ]
    )

hist = Histogram(df, values='mpg', title="Auto MPG Histogram", plot_width=400, plot_height=300, 
                 tools=[hover,'resize,reset,crosshair'])

# Histogram with default tools
hist2 = Histogram(df, values='mpg', label='cyl', color='cyl', legend='top_right',
                  title="MPG Histogram by Cylinder Count", plot_width=400, plot_height=300, 
                  tools='hover,resize,reset,crosshair')

show(hplot(hist, hist2))

output_notebook()
```
