```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import urllib.request as request
import re
from ipywidgets import interact
pd.set_option("display.max_rows",1000)
pd.set_option("display.max_columns",20)
pd.set_option("display.max_colwidth", 200)
```

```python
base_url = 'http://www.kijiji.ca'
toronto_url = 'http://www.kijiji.ca/h-city-of-toronto/1700273'
html_kijiji = request.urlopen(toronto_url)

soup_kijiji = BeautifulSoup(html_kijiji, 'lxml')
```

```python
div = soup_kijiji.find_all('a', class_='category-selected')
```

### We are going to skip the 1st 10 rows, since they are just the top level categories

```python
div_categories = div[10:]
div_categories
```

```python
categories = {}
for item in div_categories:
    categories[item.get_text()] = base_url + item['href']
```

```python
categories
```

```python
category_list = [key for key in categories.keys()]
```

```python
@interact
def kijifi_listings(category = sorted(category_list)):
    listings_list = pd.read_html(categories[category])
    dfs = listings_list[1:]
    df = pd.concat(dfs)
    return df
```

```python
url = 'http://www.kijiji.ca/b-cars-trucks/city-of-toronto/c174l1700273'
html_cars = request.urlopen(url)

soup_cars = BeautifulSoup(html_cars, 'lxml')
```

```python
#tables = soup_cars.find_all('table',  class_ = re.compile('regular-ad|top-'))
tables = soup_cars.find_all('table')
```

```python
tables[1:]
```

```python
prices = []
for table in tables[1:]:
    for row in table.find_all('td', class_='price'):
        prices.append(row.get_text().strip())
```

```python
prices
```

```python
len(prices)
```

```python
img_urls = []
for table in tables[1:]:
    for row in table.find_all('td', class_='image'):
        try:
            img_urls.append("< href='" + row.div.img['src'] + "\'>" + row.div.img['src'] + "</a>")
        except:
            img_urls.append("<a href='" + row.img['src'] + "\'>" + row.img['src'] + "</a>")
```

```python
img_urls = []
for table in tables[1:]:
    for row in table.find_all('td', class_='image'):
        try:
            img_urls.append("<img src='" + row.div.img['src'] + "'>")
        except:
            img_urls.append("<img src='" + row.img['src'] + "'>")
```

```python
img_urls
```

```python
len(img_urls)
```

```python
df = pd.DataFrame({'Price':prices, 'Image':img_urls})
```
