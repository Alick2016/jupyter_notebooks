### Read Excel file as input

```r
library(readxl)
library(choroplethr)
states_data = read_excel("d:/temp/air_mix_motor.xlsx", sheet = "Sheet1")

choropleths = list()
for (i in 2:ncol(states_data)) {
  df           = states_data[, c(1, i)]
  colnames(df) = c("region", "value")
  title        = paste0(colnames(states_data)[i])
  choropleths[[i-1]] = state_choropleth(df, title=title, legend="Claim QTY", num_colors=9)
}
```
